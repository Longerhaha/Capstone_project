### 可视化一些训练集图片，从中可以看出来图片大小是不一样的，但是我们可以使用keras中的[图片生成器](https://keras.io/preprocessing/image/),其能够将猫狗图片reshape成一样大小的图片。

```python
import matplotlib.pyplot as plt
import random
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

train_pic_num = len(train_filenames)
for i in range(15):
    plt.subplot(3, 5, i+1)
    index = random.randint(0, train_pic_num-1)
    name = train_filenames[index].split('.')[0]
    img = plt.imread('./train/'+train_filenames[index])
    plt.title(name)
    plt.imshow(img)
    plt.axis('off')
```
